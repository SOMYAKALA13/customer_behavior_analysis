```python
import pandas as pd
from sqlalchemy import create_engine, text

# ============================================
# STEP 1 - LOAD YOUR ORIGINAL DATA
# ============================================
df = pd.read_csv(r"C:\Users\Somya Kala\Downloads\customer_shopping_behavior.csv")  # Replace with your actual filename
print("✅ Data Loaded!")
print(f"Shape: {df.shape}")

# ============================================
# STEP 2 - CLEAN COLUMN NAMES
# ============================================
# Lowercase all column names
df.columns = df.columns.str.lower()

# Replace spaces with underscores
df.columns = df.columns.str.replace(' ', '_')

# Rename purchase amount column
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})

print("✅ Columns Renamed!")
print(df.columns.tolist())

# ============================================
# STEP 3 - DROP UNWANTED COLUMNS
# ============================================
df = df.drop(columns=["promo_code_used"])

print("✅ Promo Code Column Dropped!")

# ============================================
# STEP 4 - REMOVE NULL VALUES
# ============================================
df = df.dropna(subset=["customer_id"])

print(f"✅ Nulls Removed! Remaining rows: {len(df)}")

# ============================================
# STEP 5 - ADD FREQUENCY MAPPING COLUMN
# ============================================
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

print("✅ Frequency Mapping Done!")
print(df[['frequency_of_purchases', 'purchase_frequency_days']].head())

# ============================================
# STEP 6 - VERIFY DATAFRAME BEFORE PUSHING
# ============================================
print("\n✅ Final Column Names:")
print(df.columns.tolist())
print(f"\n✅ Final Shape: {df.shape}")
print("\n✅ Preview:")
print(df.head())

# ============================================
# STEP 7 - CONNECT TO MYSQL
# ============================================
engine = create_engine("mysql+pymysql://root:13082003@localhost:3306/Customer_behavior")

try:
    with engine.connect() as conn:
        result = conn.execute(text("SELECT DATABASE();"))
        print("\n✅ Connected to:", result.fetchone())
except Exception as e:
    print("❌ Connection Error:", e)

# ============================================
# STEP 8 - DROP OLD TABLE & PUSH TO MYSQL
# ============================================
try:
    # Drop old table
    with engine.connect() as conn:
        conn.execute(text("DROP TABLE IF EXISTS customer;"))
        conn.commit()
        print("✅ Old table dropped!")

    # Push updated DataFrame
    df.to_sql(
        name="customer",
        con=engine,
        if_exists="replace",
        index=False
    )
    print(f"✅ {len(df)} rows imported successfully!")

except Exception as e:
    print("❌ Import Error:", e)

# ============================================
# STEP 9 - VERIFY IN JUPYTER
# ============================================
df_verify = pd.read_sql("SELECT * FROM customer LIMIT 5", engine)
print("\n✅ MySQL Columns:")
print(df_verify.columns.tolist())
print("\n✅ Preview from MySQL:")
print(df_verify.head())
```

```python
from sqlalchemy import create_engine, text
import pandas as pd

# ============================================
# STEP 1 - APPLY AGE GROUP CHANGE
# ============================================
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

# Convert to string for MySQL compatibility
df['age_group'] = df['age_group'].astype(str)

print("✅ Age Group Column Added!")
print(df[['age', 'age_group']].head(10))
print("\nAge Group Distribution:")
print(df['age_group'].value_counts())

# ============================================
# STEP 2 - VERIFY ALL COLUMNS
# ============================================
print("\n✅ All Columns:")
print(df.columns.tolist())

# ============================================
# STEP 3 - CONNECT TO MYSQL
# ============================================
engine = create_engine("mysql+pymysql://root:13082003@localhost:3306/Customer_behavior")

try:
    with engine.connect() as conn:
        result = conn.execute(text("SELECT DATABASE();"))
        print("\n✅ Connected to:", result.fetchone())
except Exception as e:
    print("❌ Connection Error:", e)

# ============================================
# STEP 4 - DROP OLD TABLE & PUSH TO MYSQL
# ============================================
try:
    with engine.connect() as conn:
        conn.execute(text("DROP TABLE IF EXISTS customer;"))
        conn.commit()
        print("✅ Old table dropped!")

    df.to_sql(
        name="customer",
        con=engine,
        if_exists="replace",
        index=False
    )
    print(f"✅ {len(df)} rows imported with age_group column!")

except Exception as e:
    print("❌ Import Error:", e)

# ============================================
# STEP 5 - VERIFY IN JUPYTER
# ============================================
df_verify = pd.read_sql("SELECT age, age_group FROM customer LIMIT 10", engine)
print("\n✅ Age Group in MySQL:")
print(df_verify)
```
